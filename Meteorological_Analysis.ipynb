```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read csv file
df = pd.read_csv('weatherHistory.csv', usecols=['Formatted Date', 'Apparent Temperature (C)', 'Humidity'])
df
```

```python
# rename columns
df.rename(columns={'Formatted Date': 'Date',
                   'Apparent Temperature (C)': 'Apparent Temperature'}, inplace=True)
df
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
# convert 'Date' column type from 'object' to 'datetime' 
df['Date'] = pd.to_datetime(df['Date'], utc=True)
df
```

```python
df.info()
```

```python
# make 'Date' column as an index
df.set_index(df['Date'], inplace=True)
df
```

```python
# resampling data monthly and calculate mean
data = df[['Apparent Temperature', 'Humidity']].resample('MS').mean()
data
```

```python
plt.figure(figsize=(15, 3))
plt.title('Variation of Apparent Temperature vs Humidity', fontsize=25)
plt.xlabel('Graph for all months', fontsize=15)
sns.lineplot(data=data)
```
